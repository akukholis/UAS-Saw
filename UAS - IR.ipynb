```python
#tweet
tweet = [
    "This is a crazy schedule. Missing South Korea, Philippines, Singapore, the entire East Coast of Australia and New Zealand. I hope you've got more dates up your sleeve post November 22nd.",
    "no shows on the east coast of Australia means you are going to lose a lot of loyal fans. The only people who can afford to go to Perth will be the rich. Thanks for the memories. I sure won't be buying any new music that you put out",
    "a katanya band kamu jelek",
    "I'm so exited but worries at the same time, since in Indonesia we all love coldplay and it's only for one day... this is gonna be a tough war!",
    "ga ada yang mau giveaway gitu wei? Atau langsung aja bilang eh gua udah beli tiket coldplay ni buat lu gratis behh boleh banget ini mah ga akan nolaküò≠ü§öüèª",
    "NEW DATES FOR SINGAPORE AND MANY OTHER COUNTRIES PLEASEüò¢ WE MISS YALLüá∏üá¨ don't forget Philippines, Thailand, South Korea, more cities in Australia and Singapore too ahh I've been anticipating a coldplay concert in sg over the past few months but no news",
    "yg masih mikir2 mau nonton, percaya deh ini bakal jd konser terbaik kalian sepanjang masa ü•π gw nntn ini th lalu, skrg ada didepan mata berasa ga percaya banget aaa ü•πü•π tips kalo udah masuk ke antrian warnya nanti, ga usah banyak mikir",
    "guys i'm so happy for the coldplay Jakarta one but: 1) why is it only one night? we need THREE nights üò≠ 2) i thought southeast asia = jakarta, kuala lumpur, singapore, manila, bangkok? ü•πüò≠",
    "It would be very sustainable to tour the rest of Australia too while you‚Äôre here, you guys haven‚Äôt played in australia in like 7 years :((",
    "Singapore is in between Jakarta and Kuala Lumpur. The flight ticket is cheap. They can choose both of them.",
   ]
# Membuat objek TF-IDF vectorizer
vectorizer = TfidfVectorizer()

# Menghitung bobot TF-IDF pada tweet
tfidf_matrix = vectorizer.fit_transform(tweet)

# Mendapatkan nama fitur (kata)
komentar = vectorizer.get_feature_names_out()

# Mencetak bobot TF-IDF untuk setiap tweet
for i in range(len(tweet)):
    print("tweet ke", i+1)
    for j in range(len(komentar)):
        feature_name = komentar[j]
        tfidf_weight = tfidf_matrix[i, j]
        print(feature_name, ":", tfidf_weight)
    print()
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment import SentimentIntensityAnalyzer

# Inisialisasi SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Analisis sentimen untuk setiap tweet
for i in range(len(tweet)):
    print("tweet ke", i+1)
    sentiment_scores = sia.polarity_scores(tweet[i])
    compound_score = sentiment_scores['compound']

    if compound_score >= 0.05:
        print("Sentimen: Positif")
    elif compound_score <= -0.05:
        print("Sentimen: Negatif (Toksis)")
    else:
        print("Sentimen: Netral")

    print()
```
